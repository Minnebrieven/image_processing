```python
from skimage.io import imread,imshow,imsave
%matplotlib notebook
from matplotlib import pyplot as plt

im = imread('tiger.jpg')

h = [(im==v).sum() for v in range(256)]
```

```python
import numpy as np

h = np.array(h)
norm_h = h/h.sum()

plt.figure()
plt.bar(range(256), norm_h)
plt.show()
```
